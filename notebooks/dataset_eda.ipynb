```python
df.corr()
# it gives pearson correlation coefficient between the features.
#  It ranges from -1 to 1, where:
# -1 indicates a perfect negative correlation (as one variable increases, the other decreases).
# 0 indicates no correlation (the variables are independent).
# 1 indicates a perfect positive correlation (as one variable increases, the other also increases).

# formula: corr(X, Y) = cov(X, Y) / (std(X) * std(Y))

# formula for covariance: cov(X, Y) = E[(X - E[X]) * (Y - E[Y])]
```

```python
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.figure(figsize=(12, 8))
plt.show()
```

```python
plt.figure(figsize=(10,6))

scatter = plt.scatter(
    df["Longitude"],
    df["Latitude"],
    c=df["MedInc"],   # Color by income
    cmap="viridis",
    alpha=0.5
)

plt.colorbar(scatter, label="Median Income")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Income Distribution Across California")
plt.show()
```
