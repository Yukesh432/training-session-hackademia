```python
passengers = pd.DataFrame({
    "PassengerId": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"]
})

tickets = pd.DataFrame({
    "PassengerId": [1, 2],
    "Fare": [100, 50]
})
```

```python
passengers
```

```python
tickets
```

```python
# how= "left", "right", "inner", "outer"
merged_df= pd.merge(passengers, tickets, on="PassengerId", how= "left")

merged_df
```
